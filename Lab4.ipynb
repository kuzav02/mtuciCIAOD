Реализовать следующие структуры данных:
  
● <b> Стек (stack): </b>  
Операции для стека: 
<i> инициализация, проверка на пустоту, добавление нового элемента в начало, извлечение элемента из начала;  </i>  
  
● <b> Дек (двусторонняя очередь, deque):  </b>  
Операции для дека: 
<i> инициализация, проверка на пустоту, добавление нового элемента в начало, добавление нового элемента в конец, извлечение элемента из начала, извлечение элемента из конца. </i>  
  
<i> Разработать программу обработки данных, содержащихся в заранее подготовленном txt-файле, в соответствии с заданиями, применив указанную в задании структуру данных. Результат работы программы вывести на экран и сохранить в отдельном txt-файле. </i>

```python
import copy


class Stack:
    def __init__(self, stack=[]):
        self.stack = copy.copy(stack)

    def push(self, item):
        self.stack.append(item)

    def pop(self):
        if self.is_empty():
            return None
        return self.stack.pop()

    def is_empty(self):
        return len(self.stack) == 0

class Deque:
    def __init__(self, deque = []):
        self.deque = copy.copy(deque)

    def isEmpty(self):
        return self.deque == []

    def addFront(self, item):
        self.deque.append(item)

    def addRear(self, item):
        self.deque.insert(0,item)

    def removeFront(self):
        if self.isEmpty():
            return None
        return self.deque.pop()

    def removeRear(self):
        if self.isEmpty():
            return None
        return self.deque.pop(0)

    def size(self):
        return len(self.deque)
```

Реализовать алгоритм, используя три стека вместо стержней А, В, С. Информация о дисках хранится в исходном файле.

```python
def move(n, lft, mid, rgh):
    if n == 1:
        rgh.push(lft.pop())
    else:
        move(n - 1, lft, rgh, mid)
        rgh.push(lft.pop())
        move(n - 1, mid, lft, rgh)


def result(a, b, c):
    f.write("\n№3: ")
    f.write("\na: ")
    f.write(' '.join(map(str, a.stack)))
    f.write("\nb: ")
    f.write(' '.join(map(str, b.stack)))
    f.write("\nc: ")
    f.write(' '.join(map(str, c.stack)))
```

### Задание №4
Дан текстовый файл с программой на алгоритмическом языке. За один просмотр файла проверить баланс круглых скобок в тексте, используя стек.

```python
def task3_1(s) -> str:
    stack = Stack()
    for i in range(len(s)):
        if s[i] == '(':
            stack.push(s[i])
        elif len(stack.stack) != 0 and s[i] == ')':
            stack.pop()
        elif s[i] == ')':
            return "NO"
    if len(stack.stack) == 0:
        return "YES"
    else:
        return "NO"
```

### Задание №5
Дан текстовый файл с программой на алгоритмическом языке. За один просмотр файла проверить баланс квадратных скобок в тексте, используя дек.

```python
def bracket(a, disk):
    for i in disk:
        if i == '[':
            if a.isEmpty():
                a.addFront(i)
                continue
            if a.removeRear() == '[':
                a.addFront(i)
                a.addFront(i)
        if i == ']':
            if a.isEmpty():
                a.addRear(i)
                continue
            if a.removeFront() == ']':
                a.addRear(i)
                a.addRear(i)
    f.write("\n№5: " + str(a.isEmpty()))
```
