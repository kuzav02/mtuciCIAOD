## Методы сортировки

Сортировка по умолчанию

```python
def default_sort(arr1):
    arr = deepcopy(arr1)
    for i, e in enumerate(arr):
        arr[i] = sorted(e)
    #print(arr, " sort")
```

Сортировка выбором

```python
def selection(data1):
    data = deepcopy(data1)
    for i, e in enumerate(data):
        for j, e2 in enumerate(e):
            #mn, = np.where(e == min(e[j:len(e)]))   #для numpy
            mn = e.index(min(e[j:len(e)]), j)
            e[j], e[mn] = e[mn], e2
    #print(data, " selection")
```

Сортировка вставкой

```python
def insertion_sort(data1):
    data = deepcopy(data1)
    for k in range(0, len(data)):
        for i in range(1, len(data[k])): 
            value = data[k][i] 
            j = i - 1 
            while j >= 0 and value < data[k][j]: 
                data[k][j + 1] = data[k][j] 
                j -= 1 
            data[k][j + 1] = value 
    #print(data, " insertion_sort") 
```

Сортировка обменом

```python
def bubble(data1):
    data = deepcopy(data1)
    for k in range(len(data)):
        for i in range(len(data[k]) - 1):
            for j in range(len(data[k]) - 1 - i):
                if data[k][j] > data[k][j+1]:
                    data[k][j], data[k][j+1] = data[k][j+1], data[k][j]
    #print(data, " bubble")
```

Сортировка Шелла

```python
def shell(data1):
    data = deepcopy(data1)
    for k in range(len(data)):
        inc = len(data[k]) // 2
        while inc:
            for i, el in enumerate(data[k]):
                while i >= inc and data[k][i - inc] > el:
                    data[k][i] = data[k][i - inc]
                    i -= inc
                data[k][i] = el
            inc = 1 if inc == 2 else int(inc * 5.0 / 10)
    #print(data, " shell")
```

Сортировка турнирная

```python
def pyramid_tournament_sort(matrix1):
    matrix = deepcopy(matrix1)
    class Heap:
        def __init__(self, array):
            self.container = array.copy()
            self.sorted = []

        def sort(self):
            while self.container:
                n = len(self.container)
                self.container[0], self.container[n-1] = self.container[n-1], self.container[0]
                self.sorted.append(self.container[n-1])
                self.container.pop(n-1)

                self.heapify(0)

        def heapify(self, i):
            largest = i

            left = 2*i + 1
            right = 2*i + 2

            if left < len(self.container) and self.container[left] > self.container[largest]:
                largest = left
            if right < len(self.container) and self.container[right] > self.container[largest]:
                largest = right

            if largest != i:
                self.container[i], self.container[largest] = self.container[largest], self.container[i]

                self.heapify(largest)

        def build_heap(self):
            n = len(self.container)
            for i in range(n//2 - 1, -1, -1):
                self.heapify(i)
                
    i = 0
    for row in matrix:
        heap = Heap(row)
        heap.build_heap()
        heap.sort()
        
        matrix[i] = list(reversed(heap.sorted))
        
        i += 1
        
    return matrix
```

Сортировка быстрая

```python
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    elem = arr[0]
    left = list(filter(lambda x: x < elem, arr))
    center = [i for i in arr if i == elem]
    right = list(filter(lambda x: x > elem, arr))

    return quick_sort(left) + center + quick_sort(right)

def quick(arr1):
    arr = deepcopy(arr1)
    for i, e in enumerate(arr):
        arr[i] = quick_sort(e)
    #print(arr, " quick")
```

Сортировка пирамидальная

```python
def heapify(arr, n, i):
    largest = i 
    left = 2 * i + 1   
    right = 2 * i + 2   
    
    if left < n and arr[i] < arr[left]:
        largest = left
    if right < n and arr[largest] < arr[right]:
        largest = right

    if largest != i:
        arr[i],arr[largest] = arr[largest],arr[i] 
        heapify(arr, n, largest)


def heapSort(arr):
    n = len(arr)
    for k in range(len(arr)):
        for i in range(n, -1, -1):
            heapify(arr[k], n, i)

        for i in range(n-1, 0, -1):
            arr[k][i], arr[k][0] = arr[k][0], arr[k][i] 
            heapify(arr[k], i, 0)

    #print(arr, " heapSort")
```
