```python
#Бинарный поиск
import random
def binary_search_recursive(array, element: int, start: int, end: int, add: bool = False) -> int:
    if start > end:
        if add:
            if element > array[start]:
                start += 1
            return start
        else:
            return -1
    mid = (start + end) // 2
    if element == array[mid]:
        return mid
    if element < array[mid]:
        return binary_search_recursive(array, element, start, mid-1, add)
    else:
        return binary_search_recursive(array, element, mid+1, end, add)


def add(array, elem):
    i = binary_search_recursive(array, elem, 0, len(array) - 1, True)
    array.insert(i, elem)

def delete(array, elem):
    i = binary_search_recursive(array, elem, 0, len(array) - 1)
    if i != -1:
        array.pop(i)
        
        
n = 5
mas = [random.randint(-1000, 1000) for j in range(n)]
mas.sort()
print(mas)
elem1 = int(input("Введите элемент для поиска "))
elem2 = int(input("Введите элемент для вставки "))
elem3 = int(input("Введите элемент для удаления "))
print("index", binary_search_recursive(mas, elem1, 0, len(mas) - 1))
add(mas, elem2)
delete(mas, elem3)
print(mas)
```

```python
#Бинарное дерево
import random
class Node:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data

    def insert(self, data):
        if self.data:
            if data < self.data:
                if self.left is None:
                    self.left = Node(data)
                else:
                    self.left.insert(data)
            elif data > self.data:
                if self.right is None:
                    self.right = Node(data)
                else:
                    self.right.insert(data)
        else:
            self.data = data

    def delete(self, value):
        if value < self.data:
            if self.left is None:
                return str(value) + " Not Found"
            return self.left.delete(value)
        elif value > self.data:
            if self.right is None:
                return str(value) + " Not Found"
            return self.right.delete(value)
        else:
            if self.left is None and self.right is None:
                self.data = None
            elif self.left is None:
                self.data = self.right.left_del(value)
            else:
                self.data = self.left.right_del(value)

    def left_del(self, data):
        if self.left is None:
            k = self.data
            self.data = None
            return k
        else:
            return self.left.left_del(data)

    def right_del(self, data):
        if self.right is None:
            k = self.data
            self.data = None
            return k
        else:
            return self.right.right_del(data)

    def find(self, value):
        if value < self.data:
            if self.left is None:
                return str(value) + " Not Found"
            return self.left.find(value)
        elif value > self.data:
            if self.right is None:
                return str(value) + " Not Found"
            return self.right.find(value)
        else:
            print(str(self.data) + ' is found')

    def print_tree(self):
        if self.left:
            self.left.print_tree()
        if self.data is not None:
            print(self.data, end=" ")
        if self.right:
            self.right.print_tree()


array = [random.randint(-1000, 1000) for j in range(10)]
root = Node(array[0])
for i in range(1, len(array)):
    root.insert(array[i])
root.print_tree()
elem1 = int(input("\nВведите элемент для поиска "))
elem2 = int(input("Введите элемент для вставки "))
elem3 = int(input("Введите элемент для удаления "))
root.find(elem1)
root.insert(elem2)
root.delete(elem3)
root.print_tree()
```

```python
#Метод Фибоначчи
import random
def fibonacci_search(array, value, add: bool = False):
    size = len(array)
    start = -1
    f0 = 0
    f1 = 1
    f2 = 1
    while f2 < size:
        f0 = f1
        f1 = f2
        f2 = f1 + f0
    while f2 > 1:
        index = min(start + f0, size - 1)
        if add:
            if index != 0 and array[index] > value >= array[index - 1] or index == 0 and array[0] > value:
                return index
            if index != len(array) - 1 and array[index] < value <= array[index + 1] or index == len(array) - 1 and array[index] < value:
                return index + 1
        else:
            if array[index] == value:
                return index
        if array[index] < value:
            f2 = f1
            f1 = f0
            f0 = f2 - f1
            start = index
        elif array[index] > value:
            f2 = f0
            f1 -= f0
            f0 = f2 - f1
    if array[size - 1] == value:
        return size - 1
    return -1

def add(array, elem):
    i = fibonacci_search(array, elem, True)
    array.insert(i, elem)


def delete(array, elem):
    i = fibonacci_search(array, elem)
    if i != -1:
        array.pop(i)

array = [random.randint(-1000, 1000) for j in range(10)]
array.sort()
print(array)
elem1 = int(input("Введите элемент для поиска "))
elem2 = int(input("Введите элемент для вставки "))
elem3 = int(input("Введите элемент для удаления "))
print("index", fibonacci_search(array, elem1))
add(array, elem2)
delete(array, elem3)
print(array)
```

```python
#Интерполяционный
import random
def intpolsearch(values, x, add: bool = False):
    idx0 = 0
    idxn = len(values) - 1
    while (add or values[idx0] <= x <= values[idxn]) and idx0 <= idxn:
        mid = idx0 + int((float(idxn - idx0) / (values[idxn] - values[idx0])) * (x - values[idx0]))
        if add:
            if x > values[idxn]:
                return len(values)
            if x < values[idx0]:
                return 0 
            if mid != 0 and values[mid] > x >= values[mid - 1]:
                return mid
            if mid != len(values) - 1 and values[mid] <= x <= values[mid + 1]:
                return mid + 1
        else:
            if values[mid] == x:
                return mid
        if values[mid] < x:
            idx0 = mid + 1
        else:
            idxn = mid - 1
    return -1


def add(array, elem):
    i = intpolsearch(array, elem, True)
    array.insert(i, elem)


def delete(array, elem):
    i = intpolsearch(array, elem)
    if i != -1:
        array.pop(i)


array = [random.randint(0, 1000) for j in range(5)]
array.sort()
print(array)
elem1 = int(input("Введите элемент для поиска "))
elem2 = int(input("Введите элемент для вставки "))
elem3 = int(input("Введите элемент для удаления "))
print("index", intpolsearch(array, elem1))
add(array, elem2)
delete(array, elem3)
print(array)
```

```python
#Метод цепочек
import random
class Node:
    def __init__(self, data=None):
        self.data = data
        self.next = None

class Hashtable:
    def __init__(self, size):
        self.data = [Node()] * size
        self.size = size

    def hash_function(self, value, size):
        return value % size

    def put(self, value):
        hash_value = self.hash_function(value, self.size)
        if self.data[hash_value].data is None:
            self.data[hash_value].data = value
        else:
            p = self.data[hash_value]
            while p.next is not None:
                p = p.next
            p.next = Node(value)

    def get(self, value):
        hash_value = self.hash_function(value, self.size)
        if self.data[hash_value].data == value:
            return True
        else:
            p = self.data[hash_value]
            while p is not None and p.data != value:
                p = p.next
            if p is not None and p.data == value:
                return True
        return False

    def delete(self, value):
        if not self.get(value):
            return -1
        hash_value = self.hash_function(value, self.size)
        if self.data[hash_value].data == value:
            self.data[hash_value].data = None
        else:
            p = self.data[hash_value]
            pre = None
            while p is not None and p.data != value:
                pre = p
                p = p.next
            if p is None:
                return -1
            else:
                pre.next = p.next

n = 10
h = Hashtable(n)
array = set([random.randint(0, 100) for j in range(n)])
for i in array:
    h.put(i)
print(array)
# Вставка
print("Вставка")
for i in range(int(input("Введите n "))):
    h.put(int(input()))
# Удаление
print("Удаление")
for i in range(int(input("Введите n "))):
    if h.delete(int(input())) == -1:
        print('Delete Error')
# Поиск
print("Поиск")
for i in range(int(input("Введите n "))):
    print(h.get(int(input())))
```

Написать программу,  которая находит хотя бы один способ решения задач.


```python
global queen
global number
queen = [None] * 8
number = 0


def print_table():
    global number
    number += 1
    print('\n%d решение проблемы восьми ферзей \t' % number)
    for x in range(8):
        for y in range(8):
            if x == queen[y]:
                print('<q>', end=' ')
            else:
                print('<->', end=' ')
        print('\t')


def attack(row, col):
    global queen
    for i in range(col):
        offset_col = abs(i - col)
        offset_row = abs(queen[i] - row)
        if queen[i] == row or offset_row == offset_col:
            return True
    return False


def decide_position(value):
    global queen
    for i in range(8):
        if not attack(i, value):
            queen[value] = i
            if value == 7:
                print_table()
            else:
                decide_position(value + 1)


decide_position(0)
```
